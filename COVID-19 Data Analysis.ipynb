## Importing neccessary libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

### Importing Dataset on Covid-19 India case time series

```python
data = pd.read_csv('case_time_series.csv') 
data.head()
```

```python
# ‘Y’ variable stores the ‘Daily Confirmed’ corona virus cases
Y = data.iloc[61:,1].values 

# ‘R’ variable stores the ‘Daily Recovered’ corona virus cases
R = data.iloc[61:,3].values

# ‘D’ variable stores the ‘Daily Deceased’ corona virus cases
D = data.iloc[61:,5].values
```

```python
#  ‘X’ variable stores the ‘Date’ column
X = data.iloc[61:,0] 
```

### Plotting Simple Plot

```python
plt.plot(X,Y)
```

### Aesthetics
##### To have control over the aesthetics of the graph such as labels, titles, color and size we shall apply more functions as shown below:-

```python
# Creating a canvas for the graph where the first value ‘25’ is the
# width argument position and ‘8’ is the height argument position of the graph.

plt.figure(figsize=(25,8)) 


# creating an object of the axes of the graph as ‘ax’ so it becomes
# easier to implement functions.

ax = plt.axes()


# Let us create grid lines for better visualization
ax.grid(linewidth=0.4, color='#8f8f8f')


# Setting the background color of the graph as black
ax.set_facecolor("black") 

# Setting the labels for x and y axis respectively
ax.set_xlabel('\nDate',size=25,color='#4bb4f2')
ax.set_ylabel('Number of Confirmed Cases\n',size=25,color='#4bb4f2')


# Plotting the graph x vs y 
ax.plot(X,Y, color='#1F77B4',marker='o',linewidth=4,markersize=15,markeredgecolor='#035E9B')
```

```python
plt.figure(figsize=(25,8))
ax = plt.axes()
ax.grid(linewidth=0.4, color='#8f8f8f')

ax.set_facecolor("black") 

ax.set_xlabel('\nDate',size=25,
              color='#4bb4f2')
              
ax.set_ylabel('Number of Confirmed Cases\n',
               size=25,color='#4bb4f2')

ax.plot(X,Y,
        color='#1F77B4',
        marker='o',
        linewidth=4,
        markersize=15,
        markeredgecolor='#035E9B')


# with the help of xticks and yticks altering the dates and font size so that they do not overlap
plt.xticks(rotation='vertical',size='20',color='black')
plt.yticks(size=20,color='black')

# making the dashes black in color
plt.tick_params(size=20,color='black')
```

```python
plt.figure(figsize=(25,8)) 
  
ax = plt.axes() 
ax.grid(linewidth=0.4, color='#8f8f8f')  
  
ax.set_facecolor("black")  
ax.set_xlabel('\nDate',size=25,color='#4bb4f2') 
ax.set_ylabel('Number of Confirmed Cases\n', 
              size=25,color='#4bb4f2') 
  
plt.xticks(rotation='vertical',size='20',color='black') 
plt.yticks(size=20,color='black') 
plt.tick_params(size=20,color='black') 



# Annotating the graph 
# The str(j) holds the ‘Y’ variable which is Daily Confirmed.
# The XY is the coordinates where the string should be plotted. 

for i,j in zip(X,Y): 
    ax.annotate(str(j),xy=(i,j+100),color='white',size='13') 
      
ax.annotate('Second Lockdown 15th April', xy=(15.2, 860), 
            xytext=(19.9,500), color='white', size='25', 
            arrowprops=dict(color='white', linewidth=0.025)) 
  
plt.title("COVID-19 IN : Daily Confrimed\n", size=50,color='#28a9ff') 
  
ax.plot(X,Y, color='#1F77B4', marker='o', linewidth=4, 
        markersize=15, markeredgecolor='#035E9B')
```

### Plotting Transmission Pie Chart

```python
# Creating list slices based on which the pie chart will be divided and its corresponding activities
slices = [62, 142, 195]
activities = ['Travel', 'Place Visit', 'Unknown']
```

```python
# Plotting Pie chart

cols=['#4C8BE2','#00e061','#fe073a']
exp = [0.2,0.02,0.02]

plt.pie(slices,
labels=activities, 
textprops=dict(size=25,color='black'),
radius=3,
colors=cols,
autopct='%2.2f%%',
explode=exp,
shadow=True,
startangle=90)

plt.title('Transmission\n\n\n\n',color='#4fb4f2',size=40)
```

### Plotting Bar Graph

```python
## Initializing Dataset
data = pd.read_csv('district.csv')
data.head()
```

```python
# ‘re’ stores Recovered corona patients count for all districts
re=data.iloc[:30,5].values

#‘de’ stores Deceased corona patients count for all districts
de=data.iloc[:30,4].values

# ‘co’ stores Confirmed corona patients count for all districts
co=data.iloc[:30,3].values

# ‘x’ stored District names
x=list(data.iloc[:30,0])
```

```python
# Aesthetics will be same as before
plt.figure(figsize=(25,10))
ax=plt.axes()

ax.set_facecolor('black')
ax.grid(linewidth=0.4, color='#8f8f8f')


plt.xticks(rotation='vertical',
           size='20',
           color='black')

plt.yticks(size='20',color='black')


ax.set_xlabel('\nDistrict',size=25,
              color='#4bb4f2')
ax.set_ylabel('No. of cases\n',size=25,
              color='#4bb4f2')


plt.tick_params(size=20,color='white')


ax.set_title('Maharashtra District wise breakdown\n',
             size=50,color='#28a9ff')



#  Plotting bar graph
plt.bar(x,co,label='re')
plt.bar(x,re,label='re',color='green')
plt.bar(x,de,label='re',color='red')

for i,j in zip(x,co):
    ax.annotate(str(int(j)),
                xy=(i,j+3),
                color='white',
                size='15')

plt.legend(['Confirmed','Recovered','Deceased'],
           fontsize=20)
```
